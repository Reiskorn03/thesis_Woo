```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import missingno as msno   
```

```python
SUBSET_FOLDER = "/Users/chanyoungwoo/Thesis/extracted_subsets"
SUBSETS = {
    "cand_kipa_subset":     "cand_kipa_subset_ver1.csv",
    "pra_hist_subset":      "pra_hist_subset_ver1.csv",
    "stathist_kipa_subset": "stathist_kipa_subset_ver1.csv",
    "tx_ki_subset":         "tx_ki_subset_ver1.csv",
    "txf_ki_subset":        "txf_ki_subset_ver1.csv",
}

for name, fname in SUBSETS.items():
    path = os.path.join(SUBSET_FOLDER, fname)
    df = pd.read_csv(path)
    
    print(f"\n--- Missingness in {name} ({df.shape[0]} rows Ã— {df.shape[1]} cols) ---")
    
    # Missing rate in Percent
    miss_pct = df.isna().mean().mul(100).sort_values(ascending=False)
    summary = miss_pct.to_frame(name="Missing Values (%)")
    
    # Output: All Missing Values
    display(summary.head(110))
    
    # Save Missingness in seperate Column
    out_summary = os.path.join(SUBSET_FOLDER, f"{name}_missingness_summary.csv")
    summary.to_csv(out_summary, index=True)
    print(f"Saved in: {out_summary}")
    
    # Plot
    plt.figure(figsize=(10, 2))
    msno.bar(df, fontsize=8, figsize=(10,2))
    plt.title(f"Missingness in {name}")
    plt.show()
```
